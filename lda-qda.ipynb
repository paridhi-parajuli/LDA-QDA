```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import colors

cmap = colors.LinearSegmentedColormap(
    "red_blue_classes",
    {
        "red": [(0, 1, 1), (1, 0.7, 0.7)],
        "green": [(0, 0.7, 0.7), (1, 0.7, 0.7)],
        "blue": [(0, 0.7, 0.7), (1, 1, 1)],
    },
)
plt.cm.register_cmap(cmap=cmap)
```

```python
import numpy as np


def dataset_fixed_cov():
    """Generate 2 Gaussians samples with the same covariance matrix"""
    n, dim = 300, 2
    np.random.seed(0)
    C = np.array([[0.0, -0.23], [0.83, 0.23]])
    X = np.r_[
        np.dot(np.random.randn(n, dim), C),
        np.dot(np.random.randn(n, dim), C) + np.array([1, 1]),
    ]
    y = np.hstack((np.zeros(n), np.ones(n)))
    return X, y


def dataset_cov():
    """Generate 2 Gaussians samples with different covariance matrices"""
    n, dim = 300, 2
    np.random.seed(0)
    C = np.array([[0.0, -1.0], [2.5, 0.7]]) * 2.0
    X = np.r_[
        np.dot(np.random.randn(n, dim), C),
        np.dot(np.random.randn(n, dim), C.T) + np.array([1, 4]),
    ]
    y = np.hstack((np.zeros(n), np.ones(n)))
    return X, y
```

```python
from scipy import linalg


def plot_data(lda, X, y, y_pred, fig_index):
    splot = plt.subplot(2, 2, fig_index)
    if fig_index == 1:
        plt.title("Linear Discriminant Analysis")
        plt.ylabel("Data with\n fixed covariance")
    elif fig_index == 2:
        plt.title("Quadratic Discriminant Analysis")
    elif fig_index == 3:
        plt.ylabel("Data with\n varying covariances")

    tp = y == y_pred  # True Positive
    tp0, tp1 = tp[y == 0], tp[y == 1]
    X0, X1 = X[y == 0], X[y == 1]
    X0_tp, X0_fp = X0[tp0], X0[~tp0]
    X1_tp, X1_fp = X1[tp1], X1[~tp1]

    # class 0: dots
    plt.scatter(X0_tp[:, 0], X0_tp[:, 1], marker=".", color="red")
    plt.scatter(X0_fp[:, 0], X0_fp[:, 1], marker="x", s=20, color="#990000")  # dark red

    # class 1: dots
    plt.scatter(X1_tp[:, 0], X1_tp[:, 1], marker=".", color="blue")
    plt.scatter(
        X1_fp[:, 0], X1_fp[:, 1], marker="x", s=20, color="#000099"
    )  # dark blue

    # class 0 and 1 : areas
    nx, ny = 200, 100
    x_min, x_max = plt.xlim()
    y_min, y_max = plt.ylim()
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx), np.linspace(y_min, y_max, ny))
    Z = lda.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    Z = Z[:, 1].reshape(xx.shape)
    plt.pcolormesh(
        xx, yy, Z, cmap="red_blue_classes", norm=colors.Normalize(0.0, 1.0), zorder=0
    )
    plt.contour(xx, yy, Z, [0.5], linewidths=2.0, colors="white")

    # means
    plt.plot(
        lda.means_[0][0],
        lda.means_[0][1],
        "*",
        color="yellow",
        markersize=15,
        markeredgecolor="grey",
    )
    plt.plot(
        lda.means_[1][0],
        lda.means_[1][1],
        "*",
        color="yellow",
        markersize=15,
        markeredgecolor="grey",
    )

    return splot


def plot_ellipse(splot, mean, cov, color):
    v, w = linalg.eigh(cov)
    u = w[0] / linalg.norm(w[0])
    angle = np.arctan(u[1] / u[0])
    angle = 180 * angle / np.pi  # convert to degrees
    # filled Gaussian at 2 standard deviation
    ell = mpl.patches.Ellipse(
        mean,
        2 * v[0] ** 0.5,
        2 * v[1] ** 0.5,
        180 + angle,
        facecolor=color,
        edgecolor="black",
        linewidth=2,
    )
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(0.2)
    splot.add_artist(ell)
    splot.set_xticks(())
    splot.set_yticks(())


def plot_lda_cov(lda, splot):
    plot_ellipse(splot, lda.means_[0], lda.covariance_, "red")
    plot_ellipse(splot, lda.means_[1], lda.covariance_, "blue")


def plot_qda_cov(qda, splot):
    plot_ellipse(splot, qda.means_[0], qda.covariance_[0], "red")
    plot_ellipse(splot, qda.means_[1], qda.covariance_[1], "blue")
```

```python
import numpy as np
class LinearDiscriminantAnalysis:
    def __init__(self):
        self.mydict = {"class":[],"prior":[],"mean":[]}
        self.det = None
        self.covariance_=None
        self.cov_inv=None
        self.means_=None
    def fit(self,x,y):
        m=np.zeros(shape=(1,len(np.unique(y))))
        mean_all = np.mean(x, axis=0)
        cov = np.transpose(x-mean_all)@(x-mean_all)/ (len(x)-1)
        self.det = (cov[0][0] * cov[1][1]) - (cov[0][1] * cov[1][0])
        self.covariance_=cov
        self.cov_inv=cov.copy()
        self.cov_inv[0][0]= cov[1][1]
        self.cov_inv[1][1]= cov[0][0]
        self.cov_inv[0][1]= (-1)*cov[0][1] 
        self.cov_inv[1][0] = (-1)*cov[1][0]
        self.cov_inv= self.cov_inv/self.det
        if self.det<0:
            self.det=self.det*(-1)
        
        for label in np.unique(y):
            prior=len(x[y==label])/ len(x)
            mean_class = np.mean(x[y==label],axis=0)
            m=np.vstack([m, mean_class])
            self.mydict["class"].append(int(label))
            self.mydict["prior"].append(prior)
            self.mydict["mean"].append(mean_class) 
        self.means_=m[1:]
        return self

    def predict(self,x_test):
        lst=[]
        for label in self.mydict["class"]:
            m= self.mydict["mean"][label]
            xx = np.diag((x_test-np.array(m))@self.cov_inv@np.transpose(x_test-np.array(m)))
            likelihood =(1/(2*np.pi*np.sqrt(self.det))) * np.exp(-0.5*xx)*self.mydict["prior"][label]
            lst.append(list(likelihood))  
        predictions = [float(i) for i in list(np.argmax(np.transpose(np.array(lst)), axis=1))]
        
        return predictions
    
    def predict_proba(self,x):
        like=[]
        for label in self.mydict["class"]:
            m= self.mydict["mean"][label]
            xx = np.diag((x-np.array(m))@self.cov_inv@np.transpose(x-np.array(m)))
            likelihood =(1/(2*np.pi*np.sqrt(self.det))) * np.exp(-0.5*xx)*self.mydict["prior"][label]
            like.append(likelihood)
        j = np.array(like).T
        like = j/np.sum(j,axis =1).reshape(len(x),1)
        return like
        
```

```python
class QuadraticDiscriminantAnalysis:
    def __init__(self):
        self.mydict = {"class":[],"prior":[],"mean":[],
                      "cov_inv":[], "det":[]}
        self.pred_class=None
        self.means_=None
        self.covariance_=[]
    def fit(self,x,y):
        m=np.zeros(shape=(1,len(np.unique(y))))
        c=np.zeros(shape=(x.shape[1],x.shape[1]))
        for label in np.unique(y):
            x_new=x[y==label]
            prior=len(x_new)/ len(x)
            mean_class = np.mean(x_new,axis=0)
            m=np.vstack([m, mean_class])
            cov = np.transpose(x_new-mean_class)@(x_new-mean_class)/ (len(x_new)-1)
            c=np.vstack([c,cov])
            det = (cov[0][0] * cov[1][1]) - (cov[0][1] * cov[1][0])
            cov_inv=cov.copy()
            cov_inv[0][0]= cov[1][1]
            cov_inv[1][1]= cov[0][0]
            cov_inv[0][1]= (-1)*cov[0][1] 
            cov_inv[1][0] = (-1)*cov[1][0]
            cov_inv=cov_inv/det
            self.mydict["class"].append(int(label))
            self.mydict["prior"].append(prior)
            self.mydict["mean"].append(mean_class)   
            self.mydict["cov_inv"].append(cov_inv)  
            self.mydict["det"].append(det)  
            self.covariance_.append(cov)
        self.means_=m[1:]
        self.covariance_ =np.array(self.covariance_)
        return self

    def predict(self,x_test):
        lst=[]
        for label in self.mydict["class"]:
            m= self.mydict["mean"][label]
            det= self.mydict["det"][label]
            cov_inv= self.mydict["cov_inv"][label]
            xx = np.diag((x_test-np.array(m))@cov_inv@np.transpose(x_test-np.array(m)))
            likelihood =(1/(2*np.pi*np.sqrt(det))) * np.exp(-0.5*xx)*self.mydict["prior"][label]
            lst.append(list(likelihood))
        predictions = list(np.argmax(np.transpose(np.array(lst)), axis=1))
        return predictions
    
    def predict_proba(self,x):
        like=[]
        for label in self.mydict["class"]:
            m= self.mydict["mean"][label]
            det= self.mydict["det"][label]
            cov_inv= self.mydict["cov_inv"][label]
            xx = np.diag((x-np.array(m))@cov_inv@np.transpose(x-np.array(m)))
            likelihood =(1/(2*np.pi*np.sqrt(det))) * np.exp(-0.5*xx)*self.mydict["prior"][label]
            like.append(list(likelihood))
        j = np.array(like).T
        like = j/np.sum(j,axis =1).reshape(len(x),1)
        return like

```

```python
plt.figure(figsize=(10, 8), facecolor="white")
plt.suptitle(
    "Linear Discriminant Analysis vs Quadratic Discriminant Analysis",
    y=0.98,
    fontsize=15,
)


for i, (X, y) in enumerate([dataset_fixed_cov(), dataset_cov()]):
    lda = LinearDiscriminantAnalysis()
    y_pred = lda.fit(X, y).predict(X)
    splot = plot_data(lda, X, y, y_pred, fig_index=2 * i + 1)
    plot_lda_cov(lda, splot)
    plt.axis("tight")
    
    #Quadratic Discriminant Analysis
    qda = QuadraticDiscriminantAnalysis()
    y_pred = qda.fit(X, y).predict(X)
    splot = plot_data(qda, X, y, y_pred, fig_index=2 * i + 2)
    plot_qda_cov(qda, splot)
    plt.axis("tight")

plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```
